```python
NYC_Boro = shp.Reader('nycBorough.shp')
```

```python
NYC_boro_records = NYC_Boro.records()
NYC_boro_shapes = NYC_Boro.shapes()
```

```python
plot(NYC_Boro)
```

```python
# cleanTaxiData['pickup_boro'] = 'NA'
cleanTaxiData['dropoff_boro'] = 'NA'
```

```python
# Suburbs 
flag = 0
count = 0

for index, row in cleanTaxiData.iterrows():
    
    # Converting the longitude and latitude to a point object
    location = row['pickup_longitude'],row['pickup_latitude']
    point = Point(location)
    
    for j in range(len(NYC_boro_shapes)):
        # if the property latitude and longitude comes under a polygon, the correspondng suburb is recorded
        # the polygon is obtained from the shapes 
        if Polygon(NYC_boro_shapes[j].points).contains(point):
            
            cleanTaxiData.at[index, 'pickup_boro'] = NYC_boro_records[j][1]
            flag = 1
            break
    if flag == 0:
        cleanTaxiData.at[index, 'pickup_boro'] = 'NA'
        
    count += 1
    
    if(count % 10000 == 0):
        print(count)
```

```python
cleanTaxiData.to_csv('boroTaxiData_pickup_boro.csv', index = False)
```
